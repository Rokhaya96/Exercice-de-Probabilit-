```python
#Loi uniforme
#Pour la densité:

import numpy as np
import matplotlib.pyplot as plt
def densuni(x):
    #densité pour a=-1 et b=2
    if (-1<=x and x<=2):
        return(1/3)
    else:
        return(0)
    
```

```python
  #fonction de repartiotion:

def repartuni(x):
    #fonction de répartition pour a=-1 et b=2
    if x<-1:
        return(0)
    elif (-1<=x and x<=2):
        return((x+1)/3)
    else:
        return(1)
```

```python
  #representation graphique:

abs=np.linspace(-5,5,200)
y=[repartuni(x) for x in abs]
plt.grid()
plt.plot(abs,y)
plt.show()
```

```python
#Loi binomiale

import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
n, p, N = 20, 0.3, int(1e4)
B = np.random.binomial(n, p, N)
f = sps.binom.pmf(np.arange(n+1), n, p)
plt.hist(B,bins=n+1,density=1,range=(0.5,n+.5),color="white",label="loi empirique")
plt.stem(np.arange(n+1),f,"r",label="loi theorique")
plt.legend()
```

```python
#Loi exponentielle
   #densité:
from scipy.stats import expon
 #expon.pdf(x,scale=1/lambda)
abs=np.linspace(0,8,200)
plt.plot(abs,expon.pdf(abs,scale=1))
plt.plot(abs,expon.pdf(abs,scale=1/3))
plt.grid()
plt.show()
```

```python
   #fonction de repartition:
from scipy.stats import expon
 #expon.cdf(x,scale=1/lambda)
abs=np.linspace(0,8,200)
plt.plot(abs,expon.cdf(abs,scale=1))
plt.plot(abs,expon.cdf(abs,scale=1/3))
plt.grid()
plt.show()
```
